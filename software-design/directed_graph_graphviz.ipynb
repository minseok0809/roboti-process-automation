```python
f = graphviz.Digraph(filename = "output/plain organogram 2.gv")
names = ["A","B","C","D","E","F","G","H","G","H","I", "K"]
positions = ["AI Hub","Train, Valid로 구분되는 Json 파일명", 
             "Train, Valid로 구분되지 않는 Json 파일명", 
             "Train, Valid 그리고 여러 가지 Topic으로 구분하는 Json 파일명",
             "나머지 데이터셋" ]
for name, position in zip(names, positions):
     f.node(name, position)
 
#Specify edges
f.edge("A","B"); f.edge("A","C"); f.edge("A","D") #CEO to Team Leads
f.edge("B","E");
f.edge("C","기계독해"); f.edge("C","도서자료 기계독해"); f.edge("C","일반상식")
f.edge("D","웹데이터 기반 한국어 말뭉치 데이터"); f.edge("D","요약문 및 레포트 생성 데이터")
f.render(filename='aihub_dataset_category_by_file_name', directory='./', format='svg')
f
```

```python
f = graphviz.Digraph(filename = "output/plain organogram 2.gv")
names = ["A","B","C","D","E","F","G","H","G","H","I", "K"]
positions = ["AI Hub","전체 리스트를 Batch Size에 따라 분할 (len(list) >= Batch Size)", 
             "len(list)가 Batch Size 이상일 때까지 JSON 파일에서 리스트 병합"]

for name, position in zip(names, positions):
     f.node(name, position)
 
#Specify edges
f.edge("A","B"); f.edge("A","C") #CEO to Team Leads
f.edge("B","나머지 데이터셋")
f.edge("C","요약문 및 레포트 생성 데이터"); f.edge("C","법률 규정 (판결서 약관 등) 텍스트 분석 데이터"); f.edge("C","도서 자료 요약")
f.render(filename='aihub_dataset_category_by_length_list', directory='./', format='svg')
f
```
