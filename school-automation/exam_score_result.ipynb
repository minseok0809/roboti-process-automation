### Problem Score Distribution

```python
data = exam_score_df['#1a (10)']

bins = np.arange(0,15,5)  
hist, bins = np.histogram(data, bins)

accumulative_weight = []
previous = 0
for i in hist:
    previous += i
    accumulative_weight.append(previous)

plt.figure(figsize=(5, 5))
bins = np.arange(0, 15, 5) 
ax = data.plot.hist(bins=bins, width=4, xlim=(0, 10), legend=False,
                  title='#1a Score\n', color='gray') 
ax.set_xlabel('\nMark')
ax.set_ylabel('Number of Students\n')
ax.set_xticks(np.arange(5, 15, 5)-3)
ax.set_xticklabels(['0 - 5', '5 - 10'])
ax.tick_params(axis='x')

rects = ax.patches

for rect, label in zip(rects, hist):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height+0.01, label,
            ha='center', va='bottom')
    
props = dict(boxstyle='round', facecolor='white', alpha=0.5)
textstr = "Score: 10" + "\nMean: " + str(round(np.mean(exam_score_df['#1a (10)']), 4)) + "\nStd: " + str(round(np.std(exam_score_df['#1a (10)']), 4))
plt.text(0, -0.28, textstr, transform=ax.transAxes, fontsize=12, verticalalignment='bottom')

plt.savefig('#1a Score.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')   
plt.show()
```
